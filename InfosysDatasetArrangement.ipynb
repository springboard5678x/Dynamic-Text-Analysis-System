# Importing Libraries

```python
import pandas as pd
import os
```

# Function for Merging all the text files

```python
def load_imdb_data(data_dir):
    data = []
    for split in ["train","test"]:
        for sentiment in ["pos","neg"]:
            folder = os.path.join(data_dir, split, sentiment)
            label = 1 if sentiment == 'pos' else 0
            for filename in os.listdir(folder):
                if filename.startswith('.'):
                    continue
                with open(os.path.join(folder, filename), 'r', encoding='utf-8') as file:
                    review = file.read()
                    data.append({'review': review, 'sentiment': label}) 
    return pd.DataFrame(data)
df = load_imdb_data(r"C:\Users\ABIR GHOSH\Desktop\Datsets for Infosys Project\IMDB_dataset\aclImdb")
```

# Unsupervised Dataset

```python
def load_unsup_to_dataframe(unsup_dir):
    reviews = []
    for filename in os.listdir(unsup_dir):
        if filename.startswith('.'):
            continue  # Skip hidden/system files
        file_path = os.path.join(unsup_dir, filename)
        with open(file_path, 'r', encoding='utf-8') as f:
            review = f.read().strip().replace('\n', ' ')
            reviews.append(review)
    df = pd.DataFrame({'review': reviews})
    return df
df_unsup = load_unsup_to_dataframe(r"C:\Users\ABIR GHOSH\Desktop\Datsets for Infosys Project\IMDB_dataset\aclImdb\train\unsup")
df_unsup
```
